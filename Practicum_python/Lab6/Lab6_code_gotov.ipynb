```python
plt.boxplot(x=data['sqft_living'])
plt.savefig('2.png')
```

```python
fig, axs = plt.subplots(figsize=(6, 6))

numbers = list(range(1,5))
count_numbers = np.bincount(data['view'])

axs.pie(count_numbers[1:], labels=numbers, autopct='%1.1f%%', startangle=90)
axs.set_title('Оценка внешнего вида дома от 1 до 4',)
fig.savefig('3.png')
```

```python
plt.figure(figsize = (12,15))
sns.heatmap(data.corr(),annot=True,cmap="coolwarm",fmt=".2f")
plt.savefig('4.png')
```

```python
def mapper(value):
  if value < 3:
    return "bad" 
  elif value == 3: 
    return "good"
  else: 
      return "exellenet"

f = plt.figure(figsize=(20,10))
data["condition_mark"] = data["condition"].transform(mapper)
sns.countplot(x = data["floors"], hue=data["condition_mark"])
plt.savefig('5.png')
```
